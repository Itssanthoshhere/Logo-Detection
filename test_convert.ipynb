```python
# Run this in Colab where logo.h5 and LOGOS/ exist (or upload them)
!pip install -q tensorflow pillow

import tensorflow as tf, os, numpy as np
from PIL import Image

h5_path = "logo.h5"
float_tflite = "logo.tflite"
int8_tflite = "logo_int8.tflite"
REP_DIR = "LOGOS"  # change if needed
IMG_SIZE = (224,224)

# Float32 conversion
model = tf.keras.models.load_model(h5_path)
converter = tf.lite.TFLiteConverter.from_keras_model(model)
open(float_tflite, "wb").write(converter.convert())
print("Saved", float_tflite)

# INT8 conversion
def rep_gen():
    count = 0
    for root,_,files in os.walk(REP_DIR):
        for f in files:
            if count >= 300: break
            if f.lower().endswith(('.jpg','.png','.jpeg')):
                img = Image.open(os.path.join(root,f)).convert("RGB").resize(IMG_SIZE)
                a = np.array(img).astype("float32")/255.0
                a = np.expand_dims(a,0)
                yield [a]
                count += 1
        if count >= 300: break

converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.representative_dataset = rep_gen
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
converter.inference_input_type = tf.int8
converter.inference_output_type = tf.int8
open(int8_tflite, "wb").write(converter.convert())
print("Saved", int8_tflite)
```
